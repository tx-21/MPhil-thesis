```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%config InlineBackend.figure_format='retina'
params = {'mathtext.default': 'regular' }          
plt.rcParams.update(params)
sns.set_style("ticks")
# construct some data like what you have:
# lstm-sg7, lstm-2-sg7, lstm-3-sg7, lstm-4-sg7
fig, ax= plt.subplots(figsize=(6,4))
plt.suptitle('NH$_{3}$N forecasting with LSTM-sg7 with different input features')
means = np.array([0.0388,0.0379,0.0373,0.0369])
std = np.array  ([0.0003,0.0009,0.0002,0.0003])
mins = np.array ([0.0383,0.0366,0.0371,0.0365])
maxes = np.array([0.0391,0.0386,0.0377,0.0371])
ax.plot(means,c='black',linestyle='--',lw=1)
x=['Uni','Multi','Uni+pos','Multi+pos']
# create stacked errorbars:
ax.errorbar(np.arange(4), means, std, fmt='ok', lw=3)
ax.errorbar(np.arange(4), means, [means - mins, maxes - means],
             fmt='.k', ecolor='gray', lw=1)
ax.set_ylim(0.036,0.04)
ax.set_xlim(-0.3,3.3)
ax.set_xticks(np.arange(0,4,1))
ax.set_xlabel('Model input feature')
ax.set_ylabel('Test loss (MSE)')
ax.set_xticklabels(x)
plt.savefig('nh3-forecasting-comparison.png',dpi=300,bbox_inches='tight') 
```
